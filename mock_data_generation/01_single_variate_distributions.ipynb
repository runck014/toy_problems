# Single Variate Distribution

A single variate distribution is simply a probablity distribution of one random variable. They form the basis for the multivariate distibutions that we commonly use in spatial data science, for tasks like creating random points. In mathematics and statistics, you may also here single variate distibutions called *univariate* distributions.

In Python, we can use the NumPy package to create single variate distributions very quickly, which we will do in the following cells. However, we can use many different probability distributions to create our data with. A probability distribution simply shows us the probability of a variable being a certain value. In practice, we want to use whatever distribution most closely aligns with whatever we are trying to model.

### Normal Distribution
A normal distribution is the most commonly used distribution. The distribution follows a bell curve, where values a clustered in the center and they taper off on either side of the center. The distribution is centered at the mean (which is also the same as the median). About 68% of the data falls within a single standard deviation of the mean, 95% falls within two standard deviations, and over 99% falls within three standard deviations. It is also referred to as a Gaussian distribution.

To create the normal distibution, we simply need to pass in the mean and standard deviation as parameters to define how we want our distribution to look like. Then, we can pass in the size of our data that we want to model, meaning that this number of values will be modeled in our probability distribution.

```python
# Generating Data
mean_norm = 5
stdev_norm = 1

normal_dist = np.random.normal(loc=mean_norm, scale=stdev_norm, size=1000)
```

```python
# Plotting histogram
plt.hist(normal_dist, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Normal Distribution')
plt.grid(True)
plt.show()
```

### Uniform Distribution
A uniform distribution is another commonly used distribution. Instead of the probability following a bell curve, as is the case for a normal distribution, for this type of distribution, all values are equally as likely, making the distibution flat. For example, rolling a dice would be an example of a discrete uniform distribution, because the probability of rolling a 1, 2, 3, 4, 5, or 6 are all the same. In this case, we will create a continuous uniform distribution, where our values do not have to be discrete number, like the dice.

To create the uniform distibution, we simply need to pass in the minimum and maximum values of the variable as parameters to define how we want our distribution to look like. Then, we can pass in the size of our data that we want to model, just like the normal distribution.

```python
# Generating Data
min_uni = 0
max_uni = 10

uniform_dist = np.random.uniform(low=min_uni, high=max_uni, size=1000)
```

```python
# Plotting histogram
plt.hist(uniform_dist, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Uniform Distribution')
plt.grid(True)
plt.show()
```

### Exponential Distribution
Exponential distributions are the first type of distribution we have seen that are not symmetrical. We say that uniform and normal distributions are symmetrical in theory, however, that may not always be the case in reality. An exponential distribution does not follow this same logic though, as there is a heavy skew in the distribution. Typically, this distribution is used to show phenomena that has fewer large values and more small values. For example, the average time of phone calls would likely follow this distribution. Exponential distributions are always continuous, but the discrete version of the distibution is called the geometric distribution.

To create the exponential distibution, we need to pass in a parameter called the scale to define how steep or flat we want our distribution to look like. Then, we can pass in the size of our data that we want to model.

```python
# Generating Data
exponential_dist = np.random.exponential(scale=1, size=1000)
```

```python
# Plotting histogram
plt.hist(exponential_dist, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Exponential Distribution')
plt.grid(True)
plt.show()
```

### Poisson Distribution
The Poisson distribution is a special type of discrete probability distribution which is used to model the probability of a given number of events ocurring in a specific period of time. This type of distribution is used to model a series of discrete events, which are called Poisson processes. Essentially, a Poisson distribution can help us model processes where events occur at a constant rate, but the events are random and independent of one another. For example, we might use a Poisson process to model the number of car accidents along a road.

To create the Poisson distibution, we need to pass in a parameter called the lambda, which defines the mean number of events within a given interval, to define how we want our distribution to look like. Then, we can pass in the size of our data that we want to model.

```python
# Generating Data
poisson_dist = np.random.poisson(lam=5, size=1000)
```

```python
# Plotting histogram
plt.hist(poisson_dist, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Poisson Distribution')
plt.grid(True)
plt.show()
```

### Binomial Distribution
Binomial distributions are another type of discrete probability distributions. They are used to model the number of successes in a sequence of independent experiments. For example, if we were to flip a coin there would be two possible outcomes, head and tails, which could be used to represent success and failure. We would expect the number of heads (or tails) to follow this distribution.

To create the binomial distibution, we need to pass in parameters for the number of trials and the probability of success. Then, we can pass in the size of our data that we want to model.

```python
# Generating Data
binomial_dist = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
# Plotting histogram
plt.hist(binomial_dist, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Binomial Distribution')
plt.grid(True)
plt.show()
```
